```python
import sqlite3
# Connect to SQLite database (creates a new database if it doesn't exist)
conn = sqlite3.connect('api_data.db')
cursor = conn.cursor()
# Create a table for the API data
cursor.execute('''
    CREATE TABLE IF NOT EXISTS api_data (
        id INTEGER PRIMARY KEY,
        name TEXT,
        url TEXT
    )
''')
# Insert data into the table
for i in api_data:
    recipe=api_data['hits'][i]['recipe']
    cursor.execute('INSERT INTO api_data (name, url) VALUES (?, ?)', (recipe['label'], recipe['url']))
# Commit changes and close connection
conn.commit()
conn.close()
```
