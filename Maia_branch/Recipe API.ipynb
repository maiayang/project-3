```python
import sqlite3
# Connect to SQLite database (creates a new database if it doesn't exist)
conn = sqlite3.connect('recipe_data.db')
cursor = conn.cursor()

# Create a table for the API data
cursor.execute('''
    CREATE TABLE IF NOT EXISTS recipe_data (
        id INTEGER PRIMARY KEY,
        name TEXT,
        url TEXT
    )
''')
```

```python
#Verify that table has been created
res = cursor.execute("SELECT name FROM sqlite_master")
res.fetchone()
```

```python
# Insert data into the table

for i in range (0,len(api_data['hits'])):
    recipe=api_data['hits'][i]['recipe']
    cursor.execute('INSERT INTO recipe_data (name, url) VALUES (?, ?)', 
                   (recipe['label'], recipe['url']))
conn.commit()
```
