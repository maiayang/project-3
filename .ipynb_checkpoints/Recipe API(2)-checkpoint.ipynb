```python
import sqlite3
# Connect to SQLite database (creates a new database if it doesn't exist)
conn = sqlite3.connect('Resources/recipe_data.sqlite')
cursor = conn.cursor()

# Create a table for the API data
cursor.execute('''
    CREATE TABLE IF NOT EXISTS recipe_data3 (
        id INTEGER PRIMARY KEY,
        name TEXT,
        dishType TEXT,
        cuisineType TEXT,
        dietLabels TEXT,
        healthLabels TEXT,
        calories INTEGER,
        totalTime INTEGER
    )
''')
```

```python
#Verify that table has been created
res = cursor.execute("SELECT name FROM sqlite_master")
res.fetchall()
```

```python
len(api_data['hits'])
```

```python
# Insert data into the table

for i in range (0,len(api_data['hits'])):
    recipe=api_data['hits'][i]['recipe']
    cursor.execute('INSERT INTO recipe_data3 (name, dishType, cuisineType, dietLabels, healthLabels, calories, totalTime) VALUES (?, ?, ?, ?, ?, ?, ?)', 
                   (recipe['label'], recipe['dishType'][0], recipe['cuisineType'][0], recipe['dietLabels'][0], recipe['healthLabels'][0], recipe['calories'], recipe['totalTime'])
                  )
conn.commit()
```

```python
# Read table
for row in cursor.execute("SELECT * FROM recipe_data3"):
    print(row)
```

```python
# Commit changes and close connection

conn.close()
```
